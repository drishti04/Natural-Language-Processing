#TOKENIZATION

```python
AI="""Artificial intelligence refers to the simulation of human intelligence in machines that are programmed to think like humans and mimic their actions. The term may also be applied to any machine that exhibits traits associated with a human mind such as learning and problem-solving."""
from nltk.tokenize import word_tokenize
AI_tokens=word_tokenize(AI)
AI_tokens
```

#FREQUENCY OF DISTINCT WORDS


```python
from nltk.probability import FreqDist
fdist=FreqDist()
for word in AI_tokens:
  fdist[word.lower()]+=1
fdist
```

#GETTING WORDS WITH FREQUENCY TOP 10


```python
fdist_top10=fdist.most_common(10)
fdist_top10
```

```python
from nltk.tokenize import blankline_tokenize
AI_blank= blankline_tokenize(AI)
len(AI_blank)
```

#CALULATING BIGRAMS, TRIGRAMS AND NGRAMS


```python
from nltk.util import bigrams, trigrams,ngrams
String="She sells sea shells on the sea shore"
quotes_tokens=word_tokenize(String)
quotes_tokens

quotes_bigrams=list(nltk.bigrams(quotes_tokens))
quotes_bigrams

quotes_trigrams=list(nltk.trigrams(quotes_tokens))
quotes_trigrams

quotes_ngrams=list(nltk.ngrams(quotes_tokens,4))
quotes_ngrams
```

#STEMMING- it is the process of bringing the words in root form


```python
#Porter Stemmer
from nltk.stem import PorterStemmer
pst=PorterStemmer()
pst.stem("having")

words_to_stem=["give", "giving", "given","gave"]
for words in words_to_stem:
  print(words+":"+pst.stem(words))

  #Lancaster Stemmer
  from nltk.stem import LancasterStemmer
  plt=LancasterStemmer()
  for words in words_to_stem:
    print(words+":"+plt.stem(words))
    
```

#POS PARTS OF SPEECH



```python
sent="Drishti is natural when it comes to dancing"
sent_tokens=word_tokenize(sent)
sent_tokens
```

```python
for token in sent_tokens:
  print(nltk.pos_tag([token]))
```

```python
sent1="John is eating a delicious cake"
sent1_tokens=word_tokenize(sent1)
for token1 in sent1_tokens:
  print(nltk.pos_tag([token1]))
```

#NAMED ENTITY RECOGINITION- whether it's a name, city name, org name etc.



```python
from nltk import ne_chunk
ne_sent="The US President stays in the White House"
ne_tokens=word_tokenize(ne_sent)
ne_tags=nltk.pos_tag(ne_tokens)
ne_ner=ne_chunk(ne_tags)
print(ne_ner)
```
